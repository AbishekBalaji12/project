```python
import pandas as pd
import numpy as np
import re
import nltk
```

```python
df = pd.read_csv('/content/IMDB_OMDB_Kaggle_TestSet_OMDB_Detailed.csv')
df.head()
```

```python
df['Plot'][0]
```

```python
df['clean_plot'] = df['Plot'].apply(lambda x: re.sub('[^a-zA-Z]', ' ', x))
df['clean_plot'] = df['clean_plot'].apply(lambda x: re.sub('\s+',' ',x))
df['clean_plot'] = df['clean_plot'].str.lower()
df['clean_plot']
```

```python
nltk.download('punkt')
```

```python
df['clean_plot']= df['clean_plot'].apply(lambda x: nltk.word_tokenize(x))
df['clean_plot']
```

```python
 nltk.download('stopwords')
```

```python
stop_words= nltk.corpus.stopwords.words('english')
plot= []
for sentence in df['clean_plot']:
  temp = []
  for word in sentence:
    if word not in stop_words and len(word)>=3:
      temp.append(word)
  plot.append(temp)
plot
```

```python
df['clean_plot'] = plot
df['clean_plot']
```

```python
df.head()
```

```python
df['Genre']= df['Genre'].apply(lambda x: x.split(','))
df['Actors']= df['Actors'].apply(lambda x: x.split(',')[:4])
df['Director']= df['Director'].apply(lambda x: x.split(','))
```

```python
df['Actors'][0]
```

```python
def clean(sentence):
  temp = []
  for word in sentence:
    temp.append(word.lower().replace(' ',''))
  return temp
```

```python
df['Genre']= [clean(x) for x in df['Genre']]
df['Actors']= [clean(x) for x in df['Actors']]
df['Director']= [clean(x) for x in df['Director']]
```

```python
df['Actors'][0]
```

```python
columns = ['clean_plot','Genre','Actors','Director']
l = []
for i in range(len(df)):
  words = ''
  for col in columns:
    words = words + ' '.join(df[col][i]) + ' '
  l.append(words)
l
```

```python
 df['clean_input']= l
 df = df[['Title' , 'clean_input']]
 df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
tfidf = TfidfVectorizer()
features = tfidf.fit_transform(df['clean_input'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(features,features)
print(cosine_sim)
```

```python
index = pd.Series(df['Title'])
index
```

```python
def recommend_movies(title):
  movies = []
  if title not in index.values:
    return "Movie not found!"  # Handle the case where the movie isn't in the index
  idx = index[index == title].index[0]
  print(idx)
  score = pd.Series(cosine_sim[idx]).sort_values(ascending=False)
  top10 = list(score.iloc[1:11].index)
  print(top10)
  for i in top10:
    movies.append(df['Title'][i])
  return movies
```

```python
recommend_movies('The Dark Knight Rises')
```
